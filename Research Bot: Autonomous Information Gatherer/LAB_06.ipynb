```python
# Import required libraries

import requests
from getpass import getpass
from google import genai
from IPython.display import Markdown, display
```

```python
#Enter API keys securely

SERPAPI_KEY = getpass("Enter SerpAPI API Key: ")
GEMINI_KEY = getpass("Enter Gemini API Key: ")
client = genai.Client(api_key=GEMINI_KEY)

```

```python


# Take research topic from user

topic = input("Enter an AI research topic: ")
```

```python

# Search the web using SerpAPI

url = "https://serpapi.com/search"

params = {
    "engine": "google",
    "q": f"latest research trends in {topic}",
    "api_key": SERPAPI_KEY
}

response = requests.get(url, params=params)

if response.ok:
    data = response.json()
    results = data.get("organic_results", [])[:5]
    print("Results collected:", len(results))
else:
    print(f"Error: {response.status_code} - {response.reason}")
    print("Response text:", response.text)
    # Initialize data and results to empty lists to prevent further errors
    data = {}
    results = []
```

```python
#Convert search results into text

research_text = ""

for i, item in enumerate(results, start=1):
    title = item.get("title", "No title")
    link = item.get("link", "No link")
    snippet = item.get("snippet", "No description")

    research_text += f"""
Source {i}
Title: {title}
Link: {link}
Summary: {snippet}
"""
```

```python
#Summarize research findings using Gemini
from google.genai.errors import ClientError

if results:
    prompt = f"""
    You are a Research Bot.

    Topic:
    {topic}

    Web search results:
    {research_text}

    Prepare a simple research summary with these sections:
    1. Introduction
    2. Latest Trends
    3. Applications
    4. Challenges
    5. Future Scope
    6. Conclusion

    Use beginner-friendly language.
    """
    try:
        response = client.models.generate_content(
            model="gemini-2.5-flash",
            contents=prompt
        )

        display(Markdown(response.text))
    except ClientError as e:
        if "API key not valid" in str(e):
            print("ERROR: Gemini API key is invalid. Please re-run the 'Enter API keys securely' cell (Cell ID: b0t9Y-ltY7Cz) and provide a valid Gemini API key.")
        else:
            print(f"An unexpected error occurred with the Gemini API: {e}")
else:
    print("No results found. Please try another topic.")
```

```python
# Display source links
display(Markdown("## Source Links"))

for i, item in enumerate(results, start=1):
    title = item.get("title", "Source")
    link = item.get("link", "#")
    display(Markdown(f"{i}. [{title}]({link})"))
```
