```python
from langchain.agents import create_agent
from langchain.tools import tool
from langchain_google_genai import ChatGoogleGenerativeAI
from IPython.display import Markdown, display
```

```python
import os
import getpass

@tool
def split_event_tasks(user_request: str) -> str:
    """Break the college fest request into planning subtasks."""

    tasks = [
        "Venue booking",
        "Catering",
        "Publicity",
        "Scheduling",
        "Volunteer allocation",
        "Permissions",
        "Budget planning"
    ]

    text = user_request.lower()
    if "technical" in text:
        tasks.append("Projector and internet setup")

    if "cultural" in text:
        tasks.append("Stage and sound arrangement")

    result = "Main subtasks:\n"
    for i, task in enumerate(tasks, 1):
        result += f"{i}. {task}\n"

    return result
```

```python
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.3
)
```

```python
agent = create_agent(
    model=llm,
    tools=[split_event_tasks],
    system_prompt="""
You are an Event Planner Agent.

First use the available tool to split the user's request into subtasks.
Then prepare a structured college fest plan.

Give the final answer under these headings:
1. Event Summary
2. Task Breakdown
3. Suggested Plan
4. Risks and Precautions
5. Final Checklist
"""
)
```

```python
def extract_text(output):
    if isinstance(output, str):
        return output
    elif isinstance(output, list):
        text_parts = []
        for item in output:
            if isinstance(item, dict) and "text" in item:
                text_parts.append(item["text"])
        return "\n".join(text_parts)
    else:
        return str(output)
```

```python
query = """
Plan a one-day college fest for 300 students with technical events,
cultural events, food stalls, volunteer teams, and a simple budget outline.
"""

```

```python
result = agent.invoke({
    "messages": [
        {"role": "user", "content": query}
    ]
})

```

```python

final_answer = extract_text(result["messages"][-1].content)
```

```python

display(Markdown(final_answer))
```
